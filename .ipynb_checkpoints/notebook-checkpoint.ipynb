# Orion-Augmented python implementation

## Signal Acquisition and parameter estimation

We acquire the channel state information from the batch file. This file was generated from the Intel CSI tool that allows to dump the channel state information for each subcarrier, Tx and Rx antenna. 

We apply a phase correction and then we have a result a corrected version of the the angle of departure and angle of arrival jointly. Indeed, we are using a Sage algorithm in order to have robust estimates.

```python
from Bplate import *
from sage2d import *
components =1
c = csi_matrix()
file_corr='./results/2018-01-25 10:47:48/log-2018-01-25cd10:47:53.dat'
file_target='./results/2018-01-25 10:47:48/log-2018-01-25cd10:47:57.dat'
csi = c.final_matrix(file_corr, file_target)
```

We estimate the angle of departure and angle of arrival jointly. Indeed, we are using a Sage algorithm in order to have robust estimates.

```python
s = sage2d()
[Tx,Rx,Subcarrier,length]=csi.shape
dod=zeros((length))
doa=zeros((length))
for i in range(0,length):
    [beta, f1, f2, CostFunction1, CostFunction2] = s.sage(csi[:, :, 0,i], components)
    if f1.size > 0:
        dod[i]=squeeze((arcsin(abs(f1) / 0.5)*180/pi))
        doa[i]=squeeze((arcsin(abs(f2) / 0.5)*180/pi))
```

## Angle of Departure and Angle of Arrival Estimation

```python
%matplotlib inline
import matplotlib.pyplot as plt
f, axarr = plt.subplots(1, 2 ,figsize=(20,10))
plot=axarr[0].plot(transpose(dod) ,'b');
plot=axarr[1].plot(transpose(doa),'g');
l=plt.show;
```
